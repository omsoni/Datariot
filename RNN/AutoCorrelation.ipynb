```python
import pandas as pd
from matplotlib import pyplot
data = pd.read_csv('Appliances_Energy_Prediction_Dataset.csv', header=0)
applianceE = data.iloc[0:1000,0]
series = pd.Series.from_array(applianceE)
#print(applianceE)
plt = series.plot()
pyplot.ylabel('Energy Consuption')
pyplot.xlabel('Time')
pyplot.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(series)
pyplot.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(series, lags=100)
pyplot.show()
```
